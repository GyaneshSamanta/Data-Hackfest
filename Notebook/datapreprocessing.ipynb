```python
import pandas as pd

# Load the datasets
apps = pd.read_csv(r'C:\Users\mailg\OneDrive\Documents\GitHub\Data-Hackfest\Dataset\apps.tsv', delimiter='\t', encoding='utf-8')
user_history = pd.read_csv(r'C:\Users\mailg\OneDrive\Documents\GitHub\Data-Hackfest\Dataset\user_history.tsv', delimiter='\t', encoding='utf-8')
jobs = pd.read_csv(r'C:\Users\mailg\OneDrive\Documents\GitHub\Data-Hackfest\Dataset\jobs.tsv', delimiter='\t', encoding='utf-8', error_bad_lines=False)
users = pd.read_csv(r'C:\Users\mailg\OneDrive\Documents\GitHub\Data-Hackfest\Dataset\users.tsv', delimiter='\t', encoding='utf-8')
test_users = pd.read_csv(r'C:\Users\mailg\OneDrive\Documents\GitHub\Data-Hackfest\Dataset\test_users.tsv', delimiter='\t', encoding='utf-8')
popular_jobs = pd.read_csv(r'C:\Users\mailg\OneDrive\Documents\GitHub\Data-Hackfest\Dataset\popular_jobs.csv')
window_dates = pd.read_csv(r'C:\Users\mailg\OneDrive\Documents\GitHub\Data-Hackfest\Dataset\window_dates.tsv', delimiter='\t', encoding='utf-8')

# Handle Missing Values
def fill_missing_values(df):
    for column in df.columns:
        if df[column].dtype == 'object':
            df[column].fillna('Unknown', inplace=True)
        elif df[column].dtype in ['int64', 'float64']:
            df[column].fillna(df[column].median(), inplace=True)
    return df

# Handle missing values for specific columns
user_history['JobTitle'].fillna('Unknown', inplace=True)

jobs['Title'].fillna('Unknown', inplace=True)
jobs['Description'].fillna('Unknown', inplace=True)
jobs['Requirements'].fillna('Unknown', inplace=True)
jobs['City'].fillna('Unknown', inplace=True)
jobs['State'].fillna('Unknown', inplace=True)
jobs['Country'].fillna('Unknown', inplace=True)
jobs['Zip5'].fillna('00000', inplace=True)
jobs['EndDate'].fillna(jobs['StartDate'], inplace=True)  # Assuming EndDate can be same as StartDate if missing

users['State'].fillna('Unknown', inplace=True)
users['ZipCode'].fillna('00000', inplace=True)
users['Major'].fillna('Unknown', inplace=True)
users['GraduationDate'].fillna(pd.to_datetime('1900-01-01'), inplace=True)
users['TotalYearsExperience'].fillna(users['TotalYearsExperience'].median(), inplace=True)
users['CurrentlyEmployed'].fillna('Unknown', inplace=True)

popular_jobs[' JobIds'].fillna('Unknown', inplace=True)

# Date Handling
def convert_to_datetime(df, columns):
    for col in columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')
    return df

# Convert date columns to datetime
apps = convert_to_datetime(apps, ['ApplicationDate'])
jobs = convert_to_datetime(jobs, ['StartDate', 'EndDate'])
users = convert_to_datetime(users, ['GraduationDate'])
window_dates = convert_to_datetime(window_dates, ['Train Start', 'Train End / Test Start', 'Test End'])

# Verify the changes
print(apps.head())
print(user_history.head())
print(jobs.head())
print(users.head())
print(popular_jobs.head())
print(window_dates.head())
```
