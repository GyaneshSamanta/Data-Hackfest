```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Load the datasets
users = pd.read_csv(r'C:\Users\mailg\OneDrive\Documents\GitHub\Data-Hackfest\Dataset\users.tsv', delimiter='\t', encoding='utf-8')

# Handle missing values as previously discussed

# Convert GraduationDate to datetime
users['GraduationDate'] = pd.to_datetime(users['GraduationDate'], errors='coerce')

# Extract year from GraduationDate for feature engineering
users['GraduationYear'] = users['GraduationDate'].dt.year

# Encode categorical variables
le_degree = LabelEncoder()
users['DegreeType'] = le_degree.fit_transform(users['DegreeType'])

le_major = LabelEncoder()
users['Major'] = le_major.fit_transform(users['Major'].astype(str))

le_employed = LabelEncoder()
users['CurrentlyEmployed'] = le_employed.fit_transform(users['CurrentlyEmployed'])

le_managed = LabelEncoder()
users['ManagedOthers'] = le_managed.fit_transform(users['ManagedOthers'])

# Feature engineering
users['HasMasterOrHigher'] = users['DegreeType'].apply(lambda x: 1 if x in [2, 3] else 0)  # Adjust based on encoded values for Master and PhD

import datetime
current_year = datetime.datetime.now().year
users['Age'] = current_year - users['GraduationYear']

# Filter out extreme ages
users = users[(users['Age'] > 18) & (users['Age'] < 65)]

# Select relevant features
candidate_features = ['DegreeType', 'Major', 'GraduationYear', 'WorkHistoryCount', 'TotalYearsExperience', 'CurrentlyEmployed', 'ManagedOthers', 'ManagedHowMany', 'Age', 'HasMasterOrHigher']

# Drop rows with any remaining missing values in selected features
users = users.dropna(subset=candidate_features)

# Final candidate profile features
candidate_profile = users[candidate_features]

# Normalization
scaler = StandardScaler()
candidate_profile_scaled = scaler.fit_transform(candidate_profile)

# Convert back to DataFrame for easy inspection
candidate_profile_scaled = pd.DataFrame(candidate_profile_scaled, columns=candidate_features)

# Create a profile score (example: weighted sum of features)
weights = {
    'DegreeType': 1,
    'Major': 1,
    'GraduationYear': 0.5,
    'WorkHistoryCount': 2,
    'TotalYearsExperience': 2,
    'CurrentlyEmployed': 1.5,
    'ManagedOthers': 1,
    'ManagedHowMany': 1,
    'Age': 1,
    'HasMasterOrHigher': 1.5
}

candidate_profile_scaled['ProfileScore'] = candidate_profile_scaled.apply(lambda row: sum(weights[feature] * row[feature] for feature in candidate_features), axis=1)

# Print the top rows to inspect
print(candidate_profile_scaled.head())
```

```python
from sklearn.cluster import KMeans

# Number of clusters (e.g., 5)
kmeans = KMeans(n_clusters=5, random_state=42)
candidate_profile_scaled['Cluster'] = kmeans.fit_predict(candidate_profile_scaled[candidate_features])

# Assign scores based on distance to cluster centroids
candidate_profile_scaled['DistanceToCentroid'] = kmeans.transform(candidate_profile_scaled[candidate_features]).min(axis=1)

# Print the top rows to inspect
print(candidate_profile_scaled.head())
```

```python
# Rank candidates based on their profile scores
candidate_profile_scaled['Rank'] = candidate_profile_scaled['ProfileScore'].rank(ascending=False)

# Identify the top 10% of candidates
threshold = np.percentile(candidate_profile_scaled['ProfileScore'], 90)
candidate_profile_scaled['IsSuccessful'] = candidate_profile_scaled['ProfileScore'] >= threshold

# Print the top rows to inspect
print(candidate_profile_scaled.head())
print(candidate_profile_scaled['IsSuccessful'].value_counts())
```

```python
import datetime

class CandidateScorer:
    def __init__(self, csv_path):
        self.candidate_features = ['DegreeType', 'Major', 'GraduationYear', 'WorkHistoryCount', 'TotalYearsExperience', 'CurrentlyEmployed', 'ManagedOthers', 'ManagedHowMany', 'Age', 'HasMasterOrHigher']
        self.scaler = StandardScaler()
        self.kmeans = KMeans(n_clusters=5, random_state=42)
        self.le_major = LabelEncoder()
        self.weights = {
            'DegreeType': 1,
            'Major': 1,
            'GraduationYear': 0.5,
            'WorkHistoryCount': 2,
            'TotalYearsExperience': 2,
            'CurrentlyEmployed': 1.5,
            'ManagedOthers': 1,
            'ManagedHowMany': 1,
            'Age': 1,
            'HasMasterOrHigher': 1.5
        }
        self.candidate_profile_scaled = pd.read_csv(csv_path)
        self.le_major.fit(self.candidate_profile_scaled['Major'])
        self.scaler.fit(self.candidate_profile_scaled[self.candidate_features])
        self.top_10_percentile_threshold = np.percentile(self.candidate_profile_scaled['ProfileScore'], 90)
        self._fit_kmeans()

    def _fit_kmeans(self):
        self.candidate_profile_scaled['Cluster'] = self.kmeans.fit_predict(self.candidate_profile_scaled[self.candidate_features])
        self.candidate_profile_scaled['DistanceToCentroid'] = self.kmeans.transform(self.candidate_profile_scaled[self.candidate_features]).min(axis=1)

    def preprocess_input(self, degree_type, major, graduation_year, work_history_count, total_years_experience, currently_employed, managed_others, managed_how_many):
        current_year = datetime.datetime.now().year
        data = pd.DataFrame({
            'DegreeType': [degree_type],
            'Major': [self.le_major.transform([major])[0]],  # Encode the major
            'GraduationYear': [graduation_year],
            'WorkHistoryCount': [work_history_count],
            'TotalYearsExperience': [total_years_experience],
            'CurrentlyEmployed': [currently_employed],
            'ManagedOthers': [managed_others],
            'ManagedHowMany': [managed_how_many],
            'Age': [current_year - graduation_year],
            'HasMasterOrHigher': [1 if degree_type in [2, 3] else 0]
        })
        data_scaled = self.scaler.transform(data[self.candidate_features])
        data_scaled = pd.DataFrame(data_scaled, columns=data[self.candidate_features].columns)
        return data_scaled

    def calculate_profile_score(self, data_scaled):
        profile_score = sum(self.weights[feature] * data_scaled.iloc[0][feature] for feature in self.candidate_features)
        return profile_score

    def predict_success(self, profile_score):
        return profile_score >= self.top_10_percentile_threshold

    def add_candidate(self, degree_type, major, graduation_year, work_history_count, total_years_experience, currently_employed, managed_others, managed_how_many):
        data_scaled = self.preprocess_input(degree_type, major, graduation_year, work_history_count, total_years_experience, currently_employed, managed_others, managed_how_many)
        profile_score = self.calculate_profile_score(data_scaled)
        is_successful = self.predict_success(profile_score)
        return profile_score, is_successful
```
